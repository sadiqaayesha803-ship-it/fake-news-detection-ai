```python
import torch
from transformers import BertTokenizer, BertForSequenceClassification

print("⏳ Loading BERT model...")
model = BertForSequenceClassification.from_pretrained('model/bert_fakenews')
tokenizer = BertTokenizer.from_pretrained('model/bert_fakenews')

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = model.to(device)
print(f"✅ BERT model loaded on {device}!")

def predict_bert(text):
    model.eval()
    encoding = tokenizer(
        text,
        max_length=128,
        padding='max_length',
        truncation=True,
        return_tensors='pt'
    )
    input_ids = encoding['input_ids'].to(device)
    attention_mask = encoding['attention_mask'].to(device)

    with torch.no_grad():
        outputs = model(input_ids=input_ids, attention_mask=attention_mask)
        probs = torch.softmax(outputs.logits, dim=1)
        pred = torch.argmax(probs, dim=1).item()
        confidence = probs[0][pred].item() * 100

    label = "🔴 FAKE NEWS" if pred == 1 else "🟢 REAL NEWS"
    print(f"Result:     {label}")
    print(f"Confidence: {confidence:.2f}%")
    print("-" * 40)

# Test examples
print("\nTEST 1:")
predict_bert("NASA confirms water found on Mars in major discovery")

print("TEST 2:")
predict_bert("SHOCKING! Obama secretly arrested for treason by Trump")

print("TEST 3:")
predict_bert("Federal Reserve raises interest rates amid inflation concerns")
```
